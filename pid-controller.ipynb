# PID-controller

PID-controller is a control loop feedback mechanism which aims to control one value using another value as a proxy.

u(t) = Kp⋅e(t) + Ki⋅∫e(t)dt + KD⋅ddte(t)

http://en.wikipedia.org/wiki/PID_controller

```python
from random import random as r
```

```python
L = 80
P = np.concatenate([np.linspace(r(), r(), L),
                    np.linspace(r(), r(), L),
                    np.linspace(r(), r(), L),
                    np.linspace(r(), r(), L)])
T = len(P)

def plot_pid(Kp, Ki, Kd):
    u = 0.0 # start condition
    i = 0.0 # integration accumulator
    e = 0.0 # previous error
    y = []  # signal values

    for t in xrange(T):
        err = P[t] - u
        i += err
        d = err - e
        e = err
        u = u + Kp*e + Ki*i + Kd*d
        u = max(0.0, min(u, 1.0))
        y.append(u)

    x = np.linspace(1, T, T)
    plt.plot(x, P, label='reference')
    plt.plot(x, y, label='signal')
    plt.legend()
```

```python
plot_pid(0.3, 0.3, 0.3) # normal fit
```

```python
plot_pid(0.3, 0.0, 0.0) # only proportional fit
```

```python
plot_pid(0.0, 0.3, 0.0) # only integral fit
```

```python
plot_pid(0.0, 0.0, 0.3) # only derivative fit
```
